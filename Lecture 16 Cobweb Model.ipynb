# The Cobweb model (demand and supply)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 63
γ = 21/20
b = 0
β = 1
```

```python
T = 50
```

```python
ϵ = np.random.randn(T)
```

```python
p = np.zeros(T)
p[0] = 45
```

### Naive expectations

```python
for t in range(1,T):
    p[t] = (a-b)/γ - β/γ*p[t-1] - ϵ[t]/γ
```

```python
fig, ax = plt.subplots()
ax.plot(range(T),p, color='b',linewidth =3)
ax.set_xlabel('time')
ax.set_ylabel('Equilibrium price')
ax.set_title('Market price in the Cobweb model')
```
